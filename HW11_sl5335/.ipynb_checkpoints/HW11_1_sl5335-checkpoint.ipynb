```python
from __future__ import print_function
import numpy as np
import pandas as pd
import geopandas as gp
import pylab as pl
import zipfile
import os
import sys
from fiona.crs import from_epsg
import shapely
from geopy.geocoders import Nominatim
% pylab inline
```

# Acquire Census Block dataset from NYC.gov and read as geodataframe

```python
! curl -O http://www1.nyc.gov/assets/planning/download/zip/data-maps/open-data/nycb2010_16d.zip
```

```python
zip_ref = zipfile.ZipFile('nycb2010_16d.zip', 'r')
zip_ref.extractall(os.getenv('PUIDATA'))
zip_ref.close()

os.system('mv nycb2010_16d.zip ' + os.getenv('PUIDATA'))
```

```python
ct_shape = gp.read_file(os.getenv('PUIDATA') + '/nycb2010_16d/nycb2010.shp')
```

```python
ct_shape.head()
```

# Limit dataset to Brooklyn only & convert coordinates to latlong

```python
ct_brooklyn = ct_shape[ct_shape.BoroCode == '3']
ct_brooklyn.crs = from_epsg(2263)
ct_brooklyn.to_crs(epsg = 4326, inplace = True)
```

```python
ct_brooklyn.head()
```

# Create point with CUSP coordinates

```python
geolocator = Nominatim()
location = geolocator.geocode("1 Metrotech Center NYC")
```

```python
cusp = shapely.geometry.Point(location.longitude, location.latitude)
```

```python
ct_brooklyn['with_cusp'].sum() # confirm that cusp was found in one cell only
```

# Find CUSP location in Census Block

```python
# creates a true or false column depending on whether the specific tract
# include cusp
ct_brooklyn['with_cusp'] = ct_brooklyn.geometry.contains(cusp)
```

```python

cusp_tract = ct_brooklyn[ct_brooklyn.with_cusp == True]
cusp_tract.reset_index(inplace = True)
print ('CUSP is contained in {0}'.format(cusp_tract.geometry[0]))
```

```python
cusp_tract
```

# Plot Every 10 Plus CUSP tract

```python
reducted = ct_brooklyn.iloc[: :10, :]
reducted.with_cusp.sum() # check to see if cusp is included
```

```python
reducted = reducted.append(cusp_tract)
reducted.with_cusp.sum() # cusp is now included
```
